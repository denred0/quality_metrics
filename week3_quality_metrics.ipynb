```python
data = np.genfromtxt('classification.csv', delimiter=',', skip_header = 1) # загружаю данные
```

```python
data
```

```python
TP = 0 # True positive, класс относится к 1 и мы его отнесли к 1
FP = 0 # False positive, класс относится к 0 а мы его отнесли к 1
FN = 0 # False positive, класс относится к 1 а мы его отнесли к 0
TN = 0 # True negative, класс относится к 0 и мы его отнесли к 0
```

```python
# расчет

TP = 0 
FP = 0
FN = 0 
TN = 0
for i in range(len(data)):
    if data[i,0] == data[i,1] and data[i,0] == 1:
        TP += 1
    if data[i,0] == data[i,1] and data[i,0] == 0:
        TN += 1
    if data[i,0] != data[i,1] and data[i,0] == 1:
        FN += 1
    if data[i,0] != data[i,1] and data[i,0] == 0:
        FP += 1
```

```python
print('TP ' + str(TP))
print('TN ' + str(TN))
print('FN ' + str(FN))
print('FP ' + str(FP))
```

```python
from sklearn.metrics import accuracy_score
```

```python
# доля верно угаданных, посчитали библиотекой
a = accuracy_score(data[:, 0], data[:, 1])
```

```python
a
```

```python
# доля верно угаданных посчитали вручную
accuracy = (TP + TN) / (TP + TN + FN + FP)
```

```python
accuracy
```

```python
# точность, посчитали вручную
precision = TP / (TP+FP)
precision
```

```python
# точность посчитали библиотекой
from sklearn.metrics import precision_score
p = precision_score(data[:, 0], data[:, 1])
p
```

```python
# полнота, поситали вручную
recall = TP / (TP+FN)
```

```python
recall
```

```python
# полнота, посчитали библитотекой
from sklearn.metrics import recall_score
r = recall_score(data[:, 0], data[:, 1])
r
```

```python
# f-мера, посчитали библиотекой
from sklearn.metrics import f1_score
f= f1_score(data[:, 0], data[:, 1])
f
```

```python
# f-мера, посчитали вручную
fm = 2*recall*precision/ (recall + precision)
```
